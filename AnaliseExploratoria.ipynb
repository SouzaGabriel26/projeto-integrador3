```python
import pandas as pd
import numpy as np
```

### Começar a fazer a análise exploratória

```python
infos_interlagos = pd.read_csv('./Interlagos_infos.csv')
infos_interlagos
```

```python
# duração média de pit stop dos pilotos em cada corrida (em Interlagos)
mediasPilotoCorrida = infos_interlagos.groupby(['driverId', 'raceId', 'forename', 'surname']).agg({'duration': [np.mean]}).reset_index()
mediasPilotoCorrida
```

```python
# duração média de cada piloto no cirtuito de Interlagos
mediaGeralPilotos = mediasPilotoCorrida.groupby(['driverId', 'forename', 'surname']).agg({('duration', 'mean'): [np.mean, np.std, np.var]}).reset_index()
mediaGeralPilotos
```

```python
mediaGeralPilotos.isnull().sum()
```
