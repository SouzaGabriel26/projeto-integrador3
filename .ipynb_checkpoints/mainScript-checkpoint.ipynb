```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
infos_interlagos = pd.read_csv('./Interlagos_infos.csv')
infos_interlagos
```

```python
# pegando somente dados numéricos
numerical_data = infos_interlagos.groupby(['raceId', 'statusId', 'grid', 'positionOrder']).agg({'duration': [np.mean, np.std, np.var], 'stop': [np.size]}).reset_index()
# numerical_data = numerical_data.drop('raceId', axis=1)
numerical_data
```

```python
# verificando existência de valores nulos
numerical_data.isnull().sum()
```

```python
# preenchendo os valores nulos
numerical_data = numerical_data.fillna(-1)
numerical_data.isnull().sum()
```

```python
numerical_data.describe()
```

```python
# classificando os statusId's [ 0 - não terminou, 1 - terminou ]

# statusId de corridas finalizadas
finished_race = [1, 11, 12, 13, 14, 15, 16, 17, 18,
                 19, 45, 50, 128, 53, 55, 58, 88, 111,
                 112, 113, 114, 115, 116, 117, 118, 119,
                 120, 122, 123, 124, 125, 127, 133, 134]

# função anonima que retorna 1 se x estiver no array ou 0 se não.
new_status_id = numerical_data['statusId'].apply(lambda x: 1 if x in finished_race else 0)

# atualizando os dados
numerical_data['statusId'] = new_status_id


numerical_data['statusId'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(numerical_data[('duration', 'mean')], bins=10)
plt.xlabel('duração média (seconds)')
plt.ylabel('Frequência')
plt.title('Distribuição de duração média de Pit Stops')
plt.show()
```

```python
stops_count = numerical_data['stop'].value_counts()

colors = ['red' if i < 2 else 'blue' for i in range(len(stops_count))]

stops_count.plot(kind='bar', color=colors, edgecolor='black')

plt.xlabel('Quantidade de Pit Stops')
plt.ylabel('Frequência')
plt.title('Frequência de Quantidade de Pit Stops')
plt.show()
```

<h3>Correlação positiva moderada/forte, com alguns outliers</h3>
<p>Com o gráfico, podemos notar que pode haver alguma relação entre as variáveis, porém de forma moderada/forte e com muitos outliers.</p>
<p>Portanto, estamos seguindo o raciocínio de que quanto menor a posição de largada do piloto, menos a posição final.</p>

```python
# Filtrar os dados onde statusId é igual a 0
desclassificados = numerical_data[numerical_data['statusId'] == 0]

# Criar o gráfico de dispersão
plt.scatter(numerical_data['statusId'], numerical_data['positionOrder'], label='Todos')
plt.scatter(desclassificados['statusId'], desclassificados['positionOrder'], color='red', label='Desclassificados')

# Adicionar rótulos e título
plt.xlabel('statusId')
plt.ylabel('positionOrder')
plt.title('Correlação entre statusId e positionOrder')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
# naive bayes
duration_mean = numerical_data[('duration', 'mean')]

X = numerical_data[['grid', 'statusId']]
X['mean_duration'] = duration_mean

y = numerical_data['positionOrder']

# Definição dos intervalos
intervals = [(1, 3), (4, 10), (11, 24)]

# Obter os limites inferiores e superiores dos intervalos
lower_limits = [interval[0] for interval in intervals]
upper_limits = [interval[1] for interval in intervals]

# Criar os bins para a função pd.cut
bins = lower_limits + [upper_limits[-1]]  # Adicionar o último limite superior

# Categorização do positionOrder
positionRange = pd.cut(y, bins=bins, labels=[f"{interval[0]}-{interval[1]}" for interval in intervals])

# valores NaN - tratados preenchendo com uma categoria (1-3)
positionRange.isna().sum()
positionRange = positionRange.fillna('1-3')
```

```python
# Divisão dos dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, positionRange, test_size=0.2, random_state=42)
```

```python
# Criação do modelo Naive Bayes
naive_bayes = GaussianNB()

# Treinamento do modelo
naive_bayes.fit(X_train, y_train)

# Avaliação do modelo
accuracy = naive_bayes.score(X_test, y_test)
print("Accuracy:", accuracy)

# Previsão do range da posição final baseado no grid, statusId e mean_duration
y_pred = naive_bayes.predict(X_test)
y_pred
```

```python
X_test['positionOrder'] = y_pred
X_test['positionOrder'] = X_test['positionOrder'].astype('category')
X_test
```
