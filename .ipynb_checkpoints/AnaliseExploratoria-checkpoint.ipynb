```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

### Começar a fazer a análise exploratória

```python
infos_interlagos = pd.read_csv('./Interlagos_infos.csv')
infos_interlagos
```

```python
# pegando somente dados categóricos
categorical_data = infos_interlagos.groupby(['raceId', 'grid', 'positionOrder']).agg({'duration': [np.mean], 'stop': [np.size]}).reset_index()
categorical_data
```

```python
categorical_data.describe()
```

```python
for var in categorical_data[['duration', 'stop']]:
    plt.figure(figsize=(8, 6))
    plt.hist(categorical_data[var], bins=10)
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.title('Distribuição de ' + var[0])
    plt.show()
```

<p>Os Histogramas acima são representações das colunas 'duration' e 'stop', que significam o <strong>tempo médio de pitstop dos pilotos por corrida e a quantidade de pit_stops feita</strong>, respectivamente.</p>

<ul>
    <li>
        <p><strong>Sobre a 'duration':</strong> Percebe-se que o valor médio que mais se repete está entre 22 e 24 segundos.</p>       </li>
    <li>
        <p><strong>Sobre o 'stop:</strong> 2 e 3 são quantidades de pit stops mais comuns de acontecerem.</p>
    </li>
</ul>

```python
plt.figure(figsize=(8, 6))
plt.scatter(categorical_data['grid'], categorical_data['positionOrder'])
plt.xlabel('Grid')
plt.ylabel('Final position')
plt.title('Relação entre Pos. de largada e Pos. final')
plt.show()
```

```python
# Plotar gráfico de dispersão em três dimensões
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(categorical_data['grid'], categorical_data['positionOrder'], categorical_data['duration'])
ax.set_xlabel('Grid')
ax.set_ylabel('Position Order')
ax.set_zlabel('Duration')
ax.set_title('Relação entre Grid, Position Order e Duration')
plt.show()
```

<h3>Correlação positiva fraca, com muitos outliers</h3>
<p>Com o gráfico, podemos notar que pode haver alguma relação entre as variáveis, porém de forma fraca ou quase nula.</p>
<p>A relação seria entre um piloto sair numa posição boa e terminar em uma posição boa também</p>
